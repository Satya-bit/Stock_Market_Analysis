```python
#Downloading the latest data
import os
import shutil

# Create a new directory for the specific files
specific_dir = "specific_stocks"
if not os.path.exists(specific_dir):
    os.mkdir(specific_dir)

# List of specific files you want to extract
specific_files = ['GOOGL.csv', 'FB.csv', 'AAPL.csv', 'NVDA.csv', 'TSLA.csv', 'TWTR.csv']

# Source directory where the files are located (change this if needed)
source_dir = "stocks"  # or "etfs" if some of them are ETFs

def move_specific_files(files, src, dest):
    for file in files:
        src_file = os.path.join(src, file)
        if os.path.isfile(src_file):
            shutil.move(src_file, os.path.join(dest, file))
        else:
            print(f"{file} not found in {src}")

# Move the specific files to the new directory
move_specific_files(specific_files, source_dir, specific_dir)
```

**Adding few columns in the dataset**

```python
import pandas as pd
```

```python
apple= pd.read_csv('/kaggle/input/stockmarket/AAPL.csv')
google=pd.read_csv('/kaggle/input/stockmarket/GOOGL.csv')
nvidia=pd.read_csv('/kaggle/input/stockmarket/NVDA.csv')
tesla=pd.read_csv('/kaggle/input/stockmarket/TSLA.csv')
X=pd.read_csv('/kaggle/input/stockmarket/X.csv')
facebook=pd.read_csv('/kaggle/input/facebook/META.csv')
```

```python
apple.head()
```

```python
dfs=[facebook,google,nvidia,tesla,X,apple]
```

```python
for df in dfs:
    df['MA50']=df.Close.rolling(50).mean()
    df['MA200']=df.Close.rolling(200).mean()
```

```python
apple.head(50)
```

```python
for df in dfs:
    df['Previous day close price']=df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price']=df['Close'] - df['Previous day close price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price']= df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume']=df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in volume']=df['Volume'] - df['Previous day volume']
```

```python
for df in dfs:
    df['Percent change in volume']= df.Volume.pct_change()
```
